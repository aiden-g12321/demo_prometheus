## Install JAX, NumPyro, and Prometheus

1) Make a fresh conda environment.
```bash
conda create -n demo_prometheus python=3.10
conda activate demo_prometheus
```

2) Check the CUDA version.
```bash
nvidia-smi
```

3) Install JAX with the correct CUDA. e.g.
```bash
pip install -U "jax[cuda12]"
```

4) Install NumPyro.
```bash
pip install numpyro
```

5) Install scikit-sparse from conda-forge.
```bash
conda install -c conda-forge scikit-sparse
```

5) Install Prometheus.
```bash
pip install git+ssh://git@github.com/XGI-MSU/prometheus.git
```

6) While Prometheus installs its own dependencies, there are additional packages the user may want to install to interact easily with Prometheus. e.g. `enterprise_extensions`, `corner`, `matplotlib`.
